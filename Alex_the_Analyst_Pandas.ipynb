# Groupby and Aggregation

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets_Alex_The_Analyst/Flavors.csv")
```

```python
df
```

```python
df.groupby('Liked').mean(numeric_only=True)
```

```python
df1 = df.groupby('Base Flavor')
```

```python
df1.mean(numeric_only=True)
```

```python
df.groupby('Liked').count()
```

```python
df.groupby('Base Flavor').count()
```

# As we can see above the skewness is to the Chocolate

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({"Flavor Rating" : ['mean','count']})
```

```python
df.groupby('Base Flavor').describe()
```
